#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1
#(x,y) and (x,z)

plt.plot(x,y,label='(x,y)')
plt.plot(x,z,label='(x,z)')

plt.xlabel('x label')
plt.ylabel('other label')

plt.title('(x,y) vs (x,z)')
plt.legend()
plt.show
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# your code here
# Method 2
#(x,y) and (x,z)

fig,[ax1, ax2]=plt.subplots(1,2)

ax1.plot(x,y,label='x and y')
ax2.plot(x,z,label='x and z')

ax1.set_xlabel('y label')
ax1.set_ylabel('x label')
ax1.set_title('x and y', size=15)

ax2.set_xlabel('z label')
ax2.set_title('x and z', size=15)

plt.show
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
# Method 2
#(x,y) and (x,z)

fig,[ax1, ax2]=plt.subplots(1,2, figsize=(15,5))

ax1.plot(x,y,label='x and y')
ax2.plot(x,z,label='x and z')

ax1.set_xlabel('y label')
ax1.set_ylabel('x label')
ax1.set_title('x and y', size=15)

ax2.set_xlabel('z label')
ax2.set_title('x and z', size=15)

fig.suptitle('x*2 vs x**2', size=20)

plt.show
```

##### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
a=x**2
b=np.exp(x)

# your code here

fig,[ax1, ax2]=plt.subplots(1,2, figsize=(15,5))

ax1.plot(x,a,label='y=x**2')

ax2.plot(x,b,label='y=e**x')

ax1.set_xlabel('x**2')
ax1.set_ylabel('x')
ax1.set_title('y=x**2', size=15)
ax1.set_yscale('linear')

ax2.set_xlabel('e**x')
ax2.set_title('y=e**x', size=15)
ax2.set_yscale('log')

fig.suptitle('x**2 vs e**x', size=20)

plt.show
```

```python
# example of creating graphs with a for loop


from matplotlib import gridspec

kinds = ['I','II','III','IV']
states = ['dry','wet']
props = ['prop1','prop2']
T = [10,20,30,40,50]

df = pd.DataFrame(index=range(10),columns=['a','kind','state','property','T'])
df['a']=np.random.rand(10)
df['kind']=['I','I','II','III','IV','II','III','IV','II','IV']
df['state']=['dry','wet','dry','wet','dry','wet','dry','wet','dry','wet']
df['property']=['prop1','prop1','prop1','prop1','prop1','prop2','prop2','prop2','prop2','prop2']
df['T']=[10,20,30,40,50,10,20,30,40,50]

first = df[(df['state']=='dry')&(df['property']=='prop1')]
second = df[(df['state']=='wet')&(df['property']=='prop1')]
third = df[(df['state']=='dry')&(df['property']=='prop2')]
fourth = df[(df['state']=='wet')&(df['property']=='prop2')]
dfs = [first,second,third,fourth]

fig = plt.figure(figsize=(8,8.5))
gs = gridspec.GridSpec(2,2, hspace=0.4, wspace=0.3)
colours = ['r','b','g','gold'] 
symbols = ['v','v','^','^']
titles=['dry 1','wet 1','dry 2','wet 2']

for no, df in enumerate(dfs):
    ax = fig.add_subplot(gs[no])
    for i, r in enumerate(kinds):
        #print i, r
        df2 = df[df['kind']==r]
        c = colours[i]
        m = symbols[i]
        plt.scatter(df2['T'],df2['a'],c=c,s=50.0, marker=m, edgecolor='k')

    ax = plt.xlabel('T')
    ax = plt.xticks(T)
    ax = plt.ylabel('A')
    ax = plt.title(titles[no],fontsize=12,alpha=0.75)
```

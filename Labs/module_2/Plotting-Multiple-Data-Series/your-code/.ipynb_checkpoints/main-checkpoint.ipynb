# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import numpy as np

warnings.filterwarnings('ignore')
%matplotlib inline

sns.set()
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

```python
data.shape
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
data_item=data.groupby("ItemType").sum().reset_index()
data_item
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 9
fig_size[1] = 6
plt.rcParams["figure.figsize"] = fig_size
```

```python
label=data_item.ItemType
index = np.arange(len(label))
no=data_item["RetailSales"]

plt.bar(index, no)
plt.xlabel('Item Type', fontsize=15)
plt.ylabel('Retail Sales', fontsize=15)
plt.xticks(index, label, fontsize=10, rotation=30)
plt.title('Retail Sales per Item Type')
plt.show()
```

```python
label=data_item.ItemType
index = np.arange(len(label))
no=data_item["WarehouseSales"]

plt.bar(index, no)
plt.xlabel('Item Type', fontsize=15)
plt.ylabel('Warehouse Sales', fontsize=15)
plt.xticks(index, label, fontsize=10, rotation=30)
plt.title('Warehouse Sales per Item Type')
plt.show()
```

```python
label=data_item.ItemType
index = np.arange(len(label))
no=data_item["RetailTransfers"]

plt.bar(index, no)
plt.xlabel('Item Type', fontsize=15)
plt.ylabel('Retail Transfers', fontsize=15)
plt.xticks(index, label, fontsize=10, rotation=30)
plt.title('Retail Transfers per Item Type')
plt.show()
```

```python
fig, (ax1,ax2,ax3)=plt.subplots(1,3, figsize=(20,5))
label=data_item.ItemType
index = np.arange(len(label))


ax1.bar(index, data_item["RetailSales"], color="Darkorange")
ax1.set_ylabel('Retail Sales', fontsize=10)
ax1.set_xlabel('Item Type', fontsize=10)
ax1.set_title('Retail Sales per Item Type')
ax1.set_xticklabels(label, fontsize=7, rotation=30)
    
ax2.bar(index, data_item["WarehouseSales"], color='Darkblue')
ax2.set_ylabel('Warehouse Sales', fontsize=10)
ax2.set_xlabel('Item Type', fontsize=10)
ax2.set_title('Warehouse Sales per Item Type')
ax2.set_xticklabels(label, fontsize=7, rotation=30)

ax3.bar(index, data_item["RetailTransfers"], color="m")
ax3.set_ylabel('Retail Transfers', fontsize=10)
ax3.set_xlabel('Item Type', fontsize=10)
ax3.set_title('Retail Transfers per Item Type')
ax3.set_xticklabels(label, fontsize=7, rotation=30)

plt.show()
```

```python
g = sns.FacetGrid(data, col="ItemType", height=4, aspect=.5)
g.map(sns.barplot, "RetailSales", "WarehouseSales");
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
data["TotalSales"]=data["RetailSales"]+data["WarehouseSales"]
data_Total=data.groupby(["Supplier"]).sum()
data_top10=data_Total.sort_values(by=["TotalSales"], ascending=False).head(10)
data_top10
```

```python
data_top10.TotalSales.plot.barh()
```

```python
data_top10.TotalSales.plot.barh(color="r", label="Warehouse Sales")
data_top10.RetailSales.plot.barh(color="g", label="Retail Sales")
plt.xlabel("Number of Sales", labelpad=20, weight='bold', size=12)
plt.legend()
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
#fig, ax=plt.subplot(1,3)

plt.plot(data.Month, data.RetailSales)
plt.plot(data.Month, data.RetailTransfers)
plt.plot(data.Month, data.WarehouseSales)
```

```python
data_month=data.groupby("Month").mean().reset_index()
data_month
```

```python
plt.plot(data_month.Month, data_month.RetailSales, label="Retail Sales")
plt.plot(data_month.Month, data_month.RetailTransfers, label="Retail Transfers")
plt.plot(data_month.Month, data_month.WarehouseSales, label="Warehouse Sales")
plt.xlabel("Month", labelpad=20, weight='bold', size=12)
plt.ylabel("Number of sales", labelpad=20, weight='bold', size=12)
plt.legend()
```

## 4. Plot the same information as above but as a bar chart.

```python
plt.bar(data_month.Month, data_month.WarehouseSales, label="Warehouse Sales")
plt.bar(data_month.Month, data_month.RetailTransfers, label="Retail Transfers")
plt.bar(data_month.Month, data_month.RetailSales, label="Retail Sales",bottom=data_month.WarehouseSales)
plt.xlabel("Month", labelpad=20, weight='bold', size=12)
plt.ylabel("Number of sales", labelpad=20, weight='bold', size=12)
plt.legend()
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
df4 =data.groupby(['Year','Month','ItemType'])['RetailSales'].sum().reset_index().sort_values(by=['Year','Month'],ascending = True)
df4.head()
```

```python
sns.lineplot(x = df4.index , y="RetailSales", hue="ItemType",  data=df4)
```

## 6. Plot the same information as above but as a bar chart.

```python
sns.barplot(x = df4.index , y="RetailSales", hue="ItemType",  data=df4)
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*
